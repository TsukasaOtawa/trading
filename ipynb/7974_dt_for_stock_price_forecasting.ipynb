```python
import pandas as pd

ticker = 7974
#データを読み込む
processed_data = pd.read_csv(f"../csv/{ticker}_processedData.csv")
processed_data
```

```python
from datetime import datetime as dt
df_d = pd.read_csv(f"../csv/df_d/{ticker}.csv", parse_dates=True,  index_col="to_datetime", infer_datetime_format=True)
df_d
```

```python
import datetime
import matplotlib.pyplot as plt
plt.style.use('bmh')
import numpy as np
import pandas_datareader.data as pdd
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
plt.figure(figsize=(16,8))
plt.title(ticker, fontsize=18)
plt.xlabel('Date', fontsize = 18)
plt.ylabel('Close Price JPY (¥)', fontsize = 18)
plt.plot(df_d["market_capital"])
plt.show()
```

```python
financial_data = pd.DataFrame(df_d["market_capital"])
financial_data
```

```python
# 25日後の株価を予測する
future_days = 25
# 将来の「x」日を予測する関数を作成する
financial_data['prediction'] = financial_data["market_capital"].shift(-future_days)
len(df_d)
```

```python
financial_data.head(30)
financial_data.iloc[25]
```

```python
#説明変数となるデータセットを作成し出力
X = np.array(financial_data.drop(['prediction'], 1))[:-future_days]
print(X)
# len(np.array(financial_data.drop(['prediction'], axis=1))[:-future_days])
```

```python
#目的変数となるデータセットを作成し出力
y = np.array(financial_data['prediction'])[:-future_days]
print(y)
```

```python
#75％の訓練データと25％のテストデータに分割
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
print(len(x_train), len(x_test), len(y_train), len(y_test))
print(x_train[0], x_test[0])
```

```python
#決定木モデルを作成
tree = DecisionTreeRegressor().fit(x_train, y_train)
tree
```

```python
#将来を予測するデータを取得
#最後の 'x'日を除く元のデータセットからのすべての行
x_future = financial_data.drop(['prediction'], 1)[:-future_days]
x_future
```

```python
#最後の 'x'行を取得
x_future = x_future.tail(future_days) 
x_future
```

```python
#データセットをnumpy配列に変換
x_future = np.array(x_future)
print(x_future)
```

```python
#モデルツリーの予測を表示
tree_prediction = tree.predict(x_future)
print(tree_prediction)
```

```python
#決定木分析による予測モデルを作成
predictions = tree_prediction
#データをプロット
valid =  financial_data[X.shape[0]:] #228
valid = valid.drop(['prediction'], 1)
#予測価格を保持する「予測」という新しい列を作成
valid['predictions'] = predictions
valid
```

```python
plt.figure(figsize=(16,8))
plt.title(f'{ticker}')
plt.xlabel('Days',fontsize=18)
plt.ylabel('Close Price USD ($)',fontsize=18)

plt.plot(financial_data["market_capital"])
plt.plot(valid[['market_capital', 'predictions']])

plt.legend(['Train', 'Actual', 'Prediction' ], loc='lower right')
plt.show()
```
